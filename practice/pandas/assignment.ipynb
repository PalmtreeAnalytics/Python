Write a function to perform the following operation: Increase the grade of the house by 1 if the sqft of the house is greater than 0 and less than equal to 400. If the total sqft of the house is greater than 400, increase the grade of the house by 2. After creating the above function, which syntax will be using when using apply function to run the above created function in the dataset?

```python
def grade_house(house):
    sqft_living,grade = house   
    if(sqft_living > 0 and sqft_living < 400):
        return grade+1
    else:
        return grade+2

df[['sqft_living','grade']].apply(grade_house,axis=1)
```
